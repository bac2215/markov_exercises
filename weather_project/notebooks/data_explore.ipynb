```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
```

```python
df_weather_09_to_23 = pd.read_pickle('./data/df_weather_09_to_23.pickle')
```

```python
ax = df_weather_09_to_23['Temp'].hist()
ax.set_title("Histogram of Temperature (Celsius) observed in Stockholm Summers from 2009 to 2023")
ax.set_xlabel("Temperature (Celsius)")
ax.set_ylabel("Observation Frequency")
```

```python
ax = df_weather_09_to_23['Pressure'].hist()
ax.set_title("Histogram of Pressure(millibars) observed in Stockholm Summers from 2009 to 2023")
ax.set_xlabel("Pressure(millibars)")
ax.set_ylabel("Observation Frequency")
```

```python
ax = df_weather_09_to_23['Precip'].hist(bins=20)
ax.set_title("Histogram of Precipitation (cm) observed in Stockholm Summers from 2009 to 2023")
ax.set_xlabel("Precipitation (cm)")
ax.set_ylabel("Observation Frequency")
```
